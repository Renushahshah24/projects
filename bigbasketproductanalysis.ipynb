<h3> Detect and Replace Outliers with Mean</h3>

```python

def replace_outliers_with_mean(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    mean_val = data[column].mean()
    data[column] = data[column].apply(lambda x: mean_val if x < lower or x > upper else x)
    return data

for col in ['sale_price', 'market_price', 'rating']:
    bas_df = replace_outliers_with_mean(bas_df, col)
```

<h3> Create Visualizations</h3>

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(bas_df['sale_price'], kde=True, color='skyblue')
plt.title('Distribution of Sale Price')

plt.subplot(1, 2, 2)
sns.histplot(bas_df['rating'], kde=True, color='salmon')
plt.title('Distribution of Product Ratings')

plt.tight_layout()
plt.show()
```
