<h2> Box plot </h2>

```python
plt.figure(figsize=(14, 8))
for i, col in enumerate(num_column):
    plt.subplot(2, (len(num_column)+1)//2, i+1)
    sns.boxplot(y=loan_df[col])
    plt.title(f'Box Plot of {col}')
plt.tight_layout()
plt.show()
```

<h2> Bar chart </h2>

```python
categorical_cols = ['Gender','Married','Dependents','Education','Self_Employed','property_area' ]              
```

```python
categorical_cols = loan_df.select_dtypes(include=['object']).columns.to_list()[::]
categorical_cols
```

```python
categorical_cols = loan_df.select_dtypes(include='object').columns
for col in categorical_cols:
    plt.figure(figsize=(5,4))
    sns.countplot(data=loan_df, x=col, order=loan_df[col].value_counts().index,color='purple',width=0.5)
    plt.title(f'Bar Chart of {col}')
    plt.xticks(rotation=80)
    plt.show()
```

<h2> Pie plot </h2>

```python
colors = ["red","yellow"]
for col in categorical_cols:
    plt.figure(figsize=(3,8))
    loan_df[col].value_counts().plot.pie(autopct='%.2f%%', startangle=90, shadow=True,color='colors')
    plt.title(f'Pie Chart of {col}',fontdict={'weight':'bold'})
    plt.tight_layout()
    plt.legend(bbox_to_anchor=(1.4,1.1))
    plt.show()
```

<h2>: Interpretation Tips</h2>
<h3>
Histograms: Skewed income/loan values? Might need log transformation.

Box Plots: Spot extreme values in income or loan amount.

Bar Charts: Identify imbalances in categories like Gender, Education, etc.

Pie Charts: Quick glance at category proportions, e.g., approval rates.
</h3>

<h2>2.2 Bivariate Analysis </h2>

<h2> Scatter plot</h2>

```python

plt.figure(figsize=(8, 6))
sns.scatterplot(loan_df, x='ApplicantIncome', y='LoanAmount',alpha= 0.5,marker='*',color='purple')
plt.xlabel('Applicant Income')
plt.ylabel('Loan Amount')
plt.title('Applicant Income and Loan Amount')
plt.show()
```

<h2> Pair plot </h2>

```python
column1 = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']
sns.pairplot(loan_df[column1 + ['Dependents']],hue='Dependents' ,palette='husl')
plt.suptitle('Pairwise Relationships of Numeric Variables', y=1.02)
plt.show()
```

<h2> Box plot </h2>

```python

plt.figure(figsize=(5, 5))
sns.boxplot(data=loan_df, x='Education', y='ApplicantIncome', hue='Education')
plt.title('Applicant Income by Education')
plt.show()
```

<h2> Violinplot </h2>

```python
#colors = ("yellow","black","red")

plt.figure(figsize=(8, 6))
sns.violinplot(data=loan_df, x='Property_Area', y='LoanAmount', hue='Education', split=True)
plt.title('Loan Amount Distribution by Property Area & Education')
plt.show()
```

<h2>2.3 Multivariate Analysis</h2>

```python
colors = ("yellow","pink","red","purple","green","orange")
corr_matrix = loan_df.select_dtypes(include='number').corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, color=colors, fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Features')
plt.xticks(rotation=80)
plt.show()
```

<h2> Interpretation</h2>
<h3>
Values range from -1 to 1

Closer to 1: Strong positive correlation

Closer to -1: Strong negative correlation

Closer to 0: No linear correlation
</h3>

<h2>  Plot stacked bar chart </h2>


```python
column = loan_df(('Property_Area'),('Education'))
plt.plot(column.x,column.y,kind='bar', stacked=True, figsize=(8,6), colormap='viridis')
plt.title('Education by Property Area')
plt.xlabel('Property Area')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.title('Education')
plt.show()
```

```python
cross_tab = pd.crosstab(loan_df['Property_Area'], loan_df['Self_Employed'])
cross_tab.plot(kind='bar', stacked=True, figsize=(8,6), colormap='viridis')
plt.title('Self_employed by Property Area')
plt.xlabel('Property Area')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Self_employed')
plt.tight_layout()
plt.show()
```

```python
cross_tab = pd.crosstab(loan_df['Education'], loan_df['Gender'])
cross_tab.plot(kind='bar', stacked=True, figsize=(8,6), colormap='viridis')
plt.title('Education by Gender')
plt.xlabel(' Education')
plt.ylabel('Count')
plt.xticks(rotation=80)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```
